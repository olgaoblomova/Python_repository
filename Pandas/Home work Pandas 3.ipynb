–µ—Å–ª–∏ –∏—Å—Ç–æ—á–Ω–∏–∫ traffic_source —Ä–∞–≤–µ–Ω yandex –∏–ª–∏ google, —Ç–æ –≤ source_type —Å—Ç–∞–≤–∏—Ç—Å—è organic
–¥–ª—è –∏—Å—Ç–æ—á–Ω–∏–∫–æ–≤ paid –∏ email –∏–∑ –†–æ—Å—Å–∏–∏ - —Å—Ç–∞–≤–∏–º ad
–¥–ª—è –∏—Å—Ç–æ—á–Ω–∏–∫–æ–≤ paid –∏ email –Ω–µ –∏–∑ –†–æ—Å—Å–∏–∏ - —Å—Ç–∞–≤–∏–º other
–≤—Å–µ –æ—Å—Ç–∞–ª—å–Ω—ã–µ –≤–∞—Ä–∏–∞–Ω—Ç—ã –±–µ—Ä–µ–º –∏–∑ traffic_source –±–µ–∑ –∏–∑–º–µ–Ω–µ–Ω–∏–π

```python
import pandas as pd
```

```python
#–°–æ–∑–¥–∞—é –¥–∞—Ç–∞—Ñ—Ä–µ–π–º df
df = pd.read_csv('visit_log.csv', sep = ';')
df.head()
```

```python
def source_type(df):
    """
    –§—É–Ω–∫—Ü–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ —Å–æ–∑–¥–∞–≤–∞–µ–º–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞ source_type –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ –ø–æ —Å–ª–µ–¥. –ø—Ä–∞–≤–∏–ª—É:
    - –µ—Å–ª–∏ –∏—Å—Ç–æ—á–Ω–∏–∫ traffic_source —Ä–∞–≤–µ–Ω yandex –∏–ª–∏ google, —Ç–æ –≤ source_type —Å—Ç–∞–≤–∏—Ç—Å—è organic
    - –¥–ª—è –∏—Å—Ç–æ—á–Ω–∏–∫–æ–≤ paid –∏ email –∏–∑ –†–æ—Å—Å–∏–∏ - —Å—Ç–∞–≤–∏—Ç—Å—è ad
    - –¥–ª—è –∏—Å—Ç–æ—á–Ω–∏–∫–æ–≤ paid –∏ email –Ω–µ –∏–∑ –†–æ—Å—Å–∏–∏ - —Å—Ç–∞–≤–∏—Ç—Å—è other
    - –¥–ª—è –≤—Å–µ—Ö –æ—Å—Ç–∞–ª—å–Ω—ã—Ö –≤–∞—Ä–∏–∞–Ω—Ç–æ–≤ –±–µ—Ä–µ—Ç—Å—è –∑–Ω–∞—á–µ–Ω–∏–µ –∏–∑ traffic_source –±–µ–∑ –∏–∑–º–µ–Ω–µ–Ω–∏–π
    """
    if df['traffic_source'] in ['yandex','google']:
        return 'organic'
    elif df['traffic_source'] in ['paid','email']:
        if df['region']=='Russia':
            return 'ad'
        else:
            return 'other'
    else:
        return df['traffic_source']
```

```python
df['source_type'] = df.apply(source_type, axis=1)
df.head()
```

```python
#–ü—Ä–æ—Å–º–æ—Ç—Ä –∏–Ω—Ç–µ—Ä–µ—Å—É—é—â–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π 
df[(df['source_type'] != 'organic')&(df['region'] != 'Russia')&(df['traffic_source']!='yandex')&(df['traffic_source']!='google')]
```

```python
üìò–ó–∞–¥–∞–Ω–∏–µ 3

–ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Ñ–∞–π–ª —Å –æ—Ü–µ–Ω–∫–∞–º–∏ —Ñ–∏–ª—å–º–æ–≤ ml-latest-small/ratings.csv.
–ü–æ—Å—á–∏—Ç–∞–π—Ç–µ —Å—Ä–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è –∂–∏–∑–Ω–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, –∫–æ—Ç–æ—Ä—ã–µ –≤—ã—Å—Ç–∞–≤–∏–ª–∏ –±–æ–ª–µ–µ 100 –æ—Ü–µ–Ω–æ–∫.
–ü–æ–¥ –≤—Ä–µ–º–µ–Ω–µ–º –∂–∏–∑–Ω–∏ –ø–æ–Ω–∏–º–∞–µ—Ç—Å—è —Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–º –∏ –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º
—Å—Ç–æ–ª–±—Ü–∞ timestamp –¥–ª—è –¥–∞–Ω–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è userId.
```

```python
##–°–æ–∑–¥–∞—é –¥–∞—Ç–∞—Ñ—Ä–µ–π–º df_ratings
df_ratings = pd.read_csv('ml-latest-small/ratings.csv')
```

```python
df_ratings.head()
```

```python
#–°–æ–∑–¥–∞—é –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å–æ –≤—Å–µ–º–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º–∏ –∏ –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–º –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º –≤—Ä–µ–º–µ–Ω–∏
df_min_max = df_ratings[['userId', 'timestamp']].groupby('userId').agg([min, max]).reset_index()
df_min_max
```

```python
#–ü–µ—Ä–µ–∏–º–µ–Ω–æ–≤–∞–ª–∞ —Å—Ç–æ–ª–±—Ü—ã, —á—Ç–æ–±—ã –∏–∑–±–∞–≤–∏—Ç—å—Å—è –æ—Ç –∏–µ—Ä–∞—Ä—Ö–∏–∏ –≤ –Ω–∞–∑–≤–∞–Ω–∏–∏
#(–ø—Ä–∏ –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–∏ –Ω–µ–ø–µ—Ä–µ–∏–º–µ–Ω–æ–≤–∞–Ω–Ω–æ–π —Ç–∞–±–ª–∏—Ü—ã –≤—ã–¥–∞–µ—Ç—Å—è –ø—Ä–µ–¥—É–ø—Ä–µ–∂–¥–µ–Ω–∏–µ)
df_min_max.columns = ['userId', 'timestamp_min', 'timestamp_max']
df_min_max
```

```python
#–°–æ–∑–¥–∞—é –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å–æ –≤—Å–µ–º–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º–∏ –∏ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –≤—ã—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã—Ö –æ—Ü–µ–Ω–æ–∫ –ø–æ –∫–∞–∂–¥–æ–º—É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—é
df_userId = df_ratings[['userId', 'rating']].groupby('userId').count().reset_index()
df_userId.tail()
```

```python
# –°–æ–∑–¥–∞—é –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Ç–æ–ª—å–∫–æ —Å —Ç–µ–º–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º–∏, –∫–æ—Ç–æ—Ä—ã–µ –≤—ã—Å—Ç–∞–≤–∏–ª–∏ –±–æ–ª–µ–µ 100 –æ—Ü–µ–Ω–æ–∫
df_100 = df_userId.loc[df_userId['rating'] > 100]
df_100
```

```python
#–û–±—ä–µ–¥–∏–Ω—è—é —Ç–∞–±–ª–∏—Ü—É —Å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è–º–∏,–∫–æ—Ç–æ—Ä—ã–µ –≤—ã—Å—Ç–∞–≤–∏–ª–∏ –±–æ–ª–µ–µ 100 –æ—Ü–µ–Ω–æ–∫, —Å —Ç–∞–±–ª–∏—Ü–µ–π —Å–æ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–º –∏ –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–º –≤—Ä–µ–º–µ–Ω–µ–º 
df_joined = df_100.merge(df_min_max, on='userId', how='left')
#df_100.loc[pd.isnull(df_100.rating), :]
df_joined
```

```python
#–î–æ–±–∞–≤–ª—è—é –≤ –æ–±—ä–µ–¥–∏–Ω–µ–Ω–Ω—É—é —Ç–∞–±–ª–∏—Ü—É —Å—Ç–æ–ª–±—Ü—ã —Å –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–Ω—ã–º–∏ –¥–∞—Ç–∞–º–∏
df_joined['date_in'] = pd.to_datetime(df_joined['timestamp_max'], unit='s')
df_joined['date_out'] = pd.to_datetime(df_joined['timestamp_min'], unit='s')
df_joined.head()
```

```python
# –î–æ–±–∞–≤–ª—è—é —Å—Ç–æ–ª–±–µ—Ü —Å–æ –≤—Ä–µ–º–µ–Ω–µ–º –∂–∏–∑–Ω–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
df_joined['lifetime'] = df_joined.apply(lambda df: df['date_in'] - df['date_out'], axis = 1)
df_joined.head()
```

```python
#–ü–æ–¥—Å—á–∏—Ç—ã–≤–∞—é —Å—Ä–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è –∂–∏–∑–Ω–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
df_joined['lifetime'].mean()
```
