```python
#–°–æ–∑–¥–∞—é –¥–∞—Ç–∞—Ñ—Ä–µ–º df
df = pd.read_csv('power.csv')
df.head()
```

```python
#–°–æ–∑–¥–∞—é –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å–æ —Å—Ç—Ä–∞–Ω–∞–º–∏ –ü—Ä–∏–±–∞–ª—Ç–∏–∫–∏
df_baltics = df[(df['country'] == 'Estonia')|(df['country'] == 'Lithuania')|(df['country'] == 'Latvia')]
df_baltics.head()
```

```python
#–ò–∑ –¥–∞—Ç–∞—Ñ—Ä–µ–º–∞ —Å–æ —Å—Ç—Ä–∞–Ω–º–∏ –ü—Ä–∏–±–∞–ª—Ç–∏–∫–∏ –æ—Å—Ç–∞–≤–ª—è—é –¥–∞–Ω–Ω—ã–µ —É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä—è—é—â–∏–µ —É—Å–ª–æ–≤–∏—é: –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ 4, 12 –∏ 21 –∑–∞ –ø–µ—Ä–∏–æ–¥ —Å 2005 –ø–æ 2010 –≥–æ–¥–∞, –Ω–µ —É—á–∏—Ç—ã–≤–∞—è –≤ —Ä–∞—Å—á–µ—Ç–∞—Ö –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è quantity.
df_3 = df_baltics[((df_baltics['category'] == 1)|(df_baltics['category'] == 12)|(df_baltics['category'] == 21))&(df_baltics.quantity>=0)&((df_baltics['year']<=2010)&(df_baltics['year']>=2005))]
df_3.head()
```

```python
#–ü–æ–ª—É—á–∞—é —Å—É–º–º–∞—Ä–Ω–æ–µ –ø–æ—Ç—Ä–µ–±–ª–µ–Ω–∏–µ(—Å—Ç–æ–ª–±–µ—Ü 'quantity')
df_3.quantity.sum()
```
