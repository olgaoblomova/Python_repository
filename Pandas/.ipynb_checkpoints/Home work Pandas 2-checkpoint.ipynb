- –æ—Ü–µ–Ω–∫–∞ 2 –∏ –º–µ–Ω—å—à–µ - –Ω–∏–∑–∫–∏–π —Ä–µ–π—Ç–∏–Ω–≥
- –æ—Ü–µ–Ω–∫–∞ 4 –∏ –º–µ–Ω—å—à–µ - —Å—Ä–µ–¥–Ω–∏–π —Ä–µ–π—Ç–∏–Ω–≥
- –æ—Ü–µ–Ω–∫–∞ 4.5 –∏ 5 - –≤—ã—Å–æ–∫–∏–π —Ä–µ–π—Ç–∏–Ω–≥

–†–µ–∑—É–ª—å—Ç–∞—Ç –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –∑–∞–ø–∏—à–∏—Ç–µ –≤ —Å—Ç–æ–ª–±–µ—Ü class

```python
import pandas as pd
```

```python
#–°–æ–∑–¥–∞—é –¥–∞—Ç–∞—Ñ—Ä–µ–π–º df
df = pd.read_csv('ratings.csv')
#df = pd.read_csv('ml-latest-small\ratings.csv')
df.head()
```

```python
def classification_of_films(rating):
    """
    –§—É–Ω–∫—Ü–∏—è –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä—É–µ—Ç —Ñ–∏–ª—å–º—ã –ø–æ —Å–ª–µ–¥—É—é—â–∏–º –ø—Ä–∞–≤–∏–ª–∞–º:
        - –æ—Ü–µ–Ω–∫–∞ 2 –∏ –º–µ–Ω—å—à–µ - –Ω–∏–∑–∫–∏–π —Ä–µ–π—Ç–∏–Ω–≥ (low rating)
        - –æ—Ü–µ–Ω–∫–∞ 4 –∏ –º–µ–Ω—å—à–µ - —Å—Ä–µ–¥–Ω–∏–π —Ä–µ–π—Ç–∏–Ω–≥ (average rating)
        - –æ—Ü–µ–Ω–∫–∞ 4.5 –∏ 5 - –≤—ã—Å–æ–∫–∏–π —Ä–µ–π—Ç–∏–Ω–≥ (–≤—ã—Å–æ–∫–∏–π —Ä–µ–π—Ç–∏–Ω–≥)
    """
    if (rating == 4.5)|(rating == 5):
        return 'high rating'
    elif rating <= 2:
        return 'low rating'
    else:
        return 'average rating'
```

```python
# –î–æ–±–∞–≤–ª—è—é —Å—Ç–æ–ª–±–µ—Ü class
df['class'] = df['rating'].apply(classification_of_films)
df[500:520]
```

–ù–µ–æ–±—Ö–æ–¥–∏–º–æ –Ω–∞–ø–∏—Å–∞—Ç—å –≥–µ–æ-–∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä, –∫–æ—Ç–æ—Ä—ã–π –∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ–∫–µ —Å–º–æ–∂–µ—Ç –≤—ã—Å—Ç–∞–≤–∏—Ç—å –≥–µ–æ–≥—Ä–∞—Ñ–∏—á–µ—Å–∫—É—é –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–Ω–æ—Å—Ç—å –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–º—É —Ä–µ–≥–∏–æ–Ω—É. –¢. –µ. –µ—Å–ª–∏ –ø–æ–∏—Å–∫–æ–≤—ã–π –∑–∞–ø—Ä–æ—Å —Å–æ–¥–µ—Ä–∂–∏—Ç –Ω–∞–∑–≤–∞–Ω–∏–µ –≥–æ—Ä–æ–¥–∞ —Ä–µ–≥–∏–æ–Ω–∞, —Ç–æ –≤ —Å—Ç–æ–ª–±—Ü–µ ‚Äòregion‚Äô –ø–∏—à–µ—Ç—Å—è –Ω–∞–∑–≤–∞–Ω–∏–µ —ç—Ç–æ–≥–æ —Ä–µ–≥–∏–æ–Ω–∞. –ï—Å–ª–∏ –ø–æ–∏—Å–∫–æ–≤—ã–π –∑–∞–ø—Ä–æ—Å –Ω–µ —Å–æ–¥–µ—Ä–∂–∏—Ç –Ω–∞–∑–≤–∞–Ω–∏—è –≥–æ—Ä–æ–¥–∞, —Ç–æ —Å—Ç–∞–≤–∏–º ‚Äòundefined‚Äô.

'–°–µ–≤–µ—Ä–æ-–ó–∞–ø–∞–¥': ['–ø–µ—Ç–µ—Ä–±—É—Ä–≥', '–ø—Å–∫–æ–≤', '–º—É—Ä–º–∞–Ω—Å–∫'],

```python
# –°–æ–∑–¥–∞—é –¥–∞—Ç–∞—Ñ—Ä–µ–π–º
df_keywords = pd.read_csv('ml-latest-small\keywords.csv')
df_keywords.info()
```

```python
def find_city(string):
    """
    –§—É–Ω–∫—Ü–∏—è –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –Ω–∞–∑–≤–∞–Ω–∏–µ —Ä–µ–≥–∏–æ–Ω–∞, –µ—Å–ª–∏ —Å—Ç—Ä–æ–∫–∞ (–ø–æ–∏—Å–∫–æ–≤—ã–π –∑–∞–ø—Ä–æ—Å) —Å–æ–¥–µ—Ä–∂–∏—Ç –Ω–∞–∑–≤–∞–Ω–∏–µ
    –≥–æ—Ä–æ–¥–∞ —Ä–µ–≥–∏–æ–Ω–∞, –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç 'undefined'.
    """
    geo_data = {
'–¶–µ–Ω—Ç—Ä': ['–º–æ—Å–∫–≤–∞', '—Ç—É–ª–∞', '—è—Ä–æ—Å–ª–∞–≤–ª—å'],
'–°–µ–≤–µ—Ä–æ-–ó–∞–ø–∞–¥': ['–ø–µ—Ç–µ—Ä–±—É—Ä–≥', '–ø—Å–∫–æ–≤', '–º—É—Ä–º–∞–Ω—Å–∫'],
'–î–∞–ª—å–Ω–∏–π –í–æ—Å—Ç–æ–∫': ['–≤–ª–∞–¥–∏–≤–æ—Å—Ç–æ–∫', '—Å–∞—Ö–∞–ª–∏–Ω', '—Ö–∞–±–∞—Ä–æ–≤—Å–∫']
}
    for i in string.split(' '):
        if i[:-1] in [el[:-1] for el in geo_data['–¶–µ–Ω—Ç—Ä']]:
            return '–¶–µ–Ω—Ç—Ä'
        elif (i in geo_data['–°–µ–≤–µ—Ä–æ-–ó–∞–ø–∞–¥'])|(i[:-1] in geo_data['–°–µ–≤–µ—Ä–æ-–ó–∞–ø–∞–¥']):
            return '–°–µ–≤–µ—Ä–æ-–ó–∞–ø–∞–¥'
        elif (i in geo_data['–î–∞–ª—å–Ω–∏–π –í–æ—Å—Ç–æ–∫'])|(i[:-1] in geo_data['–î–∞–ª—å–Ω–∏–π –í–æ—Å—Ç–æ–∫']):
            return '–î–∞–ª—å–Ω–∏–π –í–æ—Å—Ç–æ–∫'
        else:
            continue
    return 'undefined'
```

```python
#–î–æ–±–∞–≤–ª—è—é —Å—Ç–æ–ª–±–µ—Ü 'region'
df_keywords['region'] = df_keywords['keyword'].apply(find_city)
df_keywords.tail()
```

```python
#–ü—Ä–æ—Å–º–æ—Ç—Ä –¥–∞–Ω–Ω—ã—Ö, –≤ –∫–æ—Ç–æ—Ä—ã—Ö –≤ —Å—Ç–æ–ª–±—Ü–µ 'region' –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –∑–Ω–∞—á–µ–Ω–∏–µ –æ—Ç–ª–∏—á–Ω–æ–µ –æ—Ç'undefined'
df_keywords[df_keywords['region'] != 'undefined']
```

–ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é production_year, –∫–æ—Ç–æ—Ä–∞—è –∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ–∫–µ –∏–∑ –Ω–∞–∑–≤–∞–Ω–∏—è —Ñ–∏–ª—å–º–∞ –≤—ã—Å—Ç–∞–≤–ª—è–µ—Ç –≥–æ–¥ –≤—ã–ø—É—Å–∫–∞. –ù–µ –≤—Å–µ –Ω–∞–∑–≤–∞–Ω–∏—è —Ñ–∏–ª—å–º–æ–≤ —Å–æ–¥–µ—Ä–∂–∞—Ç –≥–æ–¥ –≤—ã–ø—É—Å–∫–∞ –≤ –æ–¥–∏–Ω–∞–∫–æ–≤–æ–º —Ñ–æ—Ä–º–∞—Ç–µ, –ø–æ—ç—Ç–æ–º—É –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Å–ª–µ–¥—É—é—â–∏–π –∞–ª–≥–æ—Ä–∏—Ç–º:

–¥–ª—è –∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ–∫–∏ –ø—Ä–æ–π–¥–∏—Ç–µ –ø–æ –≤—Å–µ–º –≥–æ–¥–∞–º —Å–ø–∏—Å–∫–∞ years

–µ—Å–ª–∏ –Ω–∏ –æ–¥–∏–Ω –∏–∑ –Ω–æ–º–µ—Ä–æ–≤ –≥–æ–¥–∞ —Å–ø–∏—Å–∫–∞ years –Ω–µ –≤—Å—Ç—Ä–µ—Ç–∏–ª—Å—è –≤ –Ω–∞–∑–≤–∞–Ω–∏–∏ —Ñ–∏–ª—å–º–∞, —Ç–æ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç—Å—è 1900 –≥–æ–¥
–ó–∞–ø–∏—à–∏—Ç–µ –≥–æ–¥ –≤—ã–ø—É—Å–∫–∞ —Ñ–∏–ª—å–º–∞ –ø–æ –∞–ª–≥–æ—Ä–∏—Ç–º—É –ø—É–Ω–∫—Ç–∞ 2 –≤ –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü ‚Äòyear‚Äô

–ü–æ—Å—á–∏—Ç–∞–π—Ç–µ —Å—Ä–µ–¥–Ω–∏–π —Ä–µ–π—Ç–∏–Ω–≥ –≤—Å–µ—Ö —Ñ–∏–ª—å–º–æ–≤ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–∞ ‚Äòyear‚Äô –∏ –æ—Ç—Å–æ—Ä—Ç–∏—Ä—É–π—Ç–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø–æ —É–±—ã–≤–∞–Ω–∏—é —Ä–µ–π—Ç–∏–Ω–≥–∞

```python
df = pd.read_csv('ratings.csv')
df.head()
```

```python
df_movies = pd.read_csv('ml-latest-small\movies.csv')
df_movies[1000:1020]
```

```python
years = [x for x in range(1950,2022)]
print(years)
print(len(years))
```

```python
def production_year(string):
    """
    –§—É–Ω–∫—Ü–∏—è –∫–∞–∂–¥–æ–π —Å—Ç—Ä–æ–∫–µ –∏–∑ –Ω–∞–∑–≤–∞–Ω–∏—è —Ñ–∏–ª—å–º–∞ –≤—ã—Å—Ç–∞–≤–ª—è–µ—Ç –≥–æ–¥ –≤—ã–ø—É—Å–∫–∞
    """
    for i in years:
        if str(i) in string:
            return str(i)
    return '1900'
```

```python
#–î–æ–±–∞–≤–ª—è—é —Å—Ç–æ–ª–±–µ—Ü 'year' –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º df_movies
df_movies['year'] = df_movies['title'].apply(production_year)
df_movies.head()
```

```python
# –û–±—ä–µ–¥–∏–Ω—è—é —Ç–∞–±–ª–∏—Ü—É —Å —Ä–µ–π—Ç–∏–Ω–≥–∞–º–∏ –∏ –Ω–∞–∑–≤–∞–Ω–∏—è–º–∏ —Ñ–∏–ª—å–º–æ–≤ –ø–æ —Å—Ç–æ–ª–±—Ü—É 'movieId'
df_joined = df.merge(df_movies, on='movieId', how='left')
df_joined.head()
```

```python
#–ì—Ä—É–ø–ø–∏—Ä—É—é —Å—Ç–æ–ª–±—Ü—ã –ø–æ –≥–æ–¥—É –∏ –Ω–∞—Ö–æ–∂—É —Å—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è
df_mean = df_joined.groupby('year').mean().reset_index()
df_mean.head(25)
```

```python
# –°–æ—Ä—Ç–∏—Ä—É—é —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ø–æ —É–±—ã–≤–∞–Ω–∏—é —Ä–µ–π—Ç–∏–Ω–≥–∞. –ü–æ—Ö–æ–∂–µ, —Ä–∞–Ω—å—à–µ —Å–Ω–∏–º–∞–ª–∏ –Ω–∞—Å—Ç–æ—è—â–µ–µ –∫–∏–Ω–æ, –Ω–µ —Ç–æ —á—Ç–æ —Å–µ–π—á–∞—Å..
df_mean.sort_values('rating', ascending=False, inplace=True)
df_mean[['rating', 'year']].head()
```
