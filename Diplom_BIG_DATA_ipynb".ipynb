##Средняя оценка по жанрам


```python
from pyspark.sql import Row 
from pyspark import SparkContext
sc =SparkContext.getOrCreate()
from pyspark.sql.types import StructType
from pyspark.sql.types import StructField
from pyspark.sql.types import StringType
from pyspark.sql.types import *
```

```python
def average_raiting(d):
  """
  Функция подсчитывает среднюю оценку для каждого жанра из списка list_genres
  """
  schema = StructType([StructField("genres", StringType(), True),
                      StructField("average_raiting", IntegerType(), False)]
                     )
  df_test = spark.createDataFrame([], schema)
  list_genres = ['unknown ',
  'Action ',
  'Adventure',
  'Animation',
  "Children's",
  'Comedy',
  'Crime',
  'Documentary',
  'Drama',
  'Fantasy',
  'Film-Noir',
  'Horror',
  'Musical',
  'Mystery',
  'Romance',
  'Sci-Fi',
  'Thriller',
  'War',
  'Western']
  row1 = Row('genres', 'average_raiting')
  for i in range(len(list_genres)):
      s = d.groupBy(list_genres[i]).mean('avg(rating)')
      s = s.where(s[list_genres[i]] == 1).collect()[0][1]
      df_avr_rait = sc.parallelize([row1(list_genres[i], s)]).toDF()
      df_test = df_test.union(df_avr_rait)
  return df_test.show()
```

Вывожу среднюю среднюю оценку для каждого жанра

```python
average_raiting(df_sum)
```
