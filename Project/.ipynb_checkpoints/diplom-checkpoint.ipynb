## ПРОВЕРКА ГИПОТЕЗЫ, что сотрудники с высоким окладом проводят на работе больше времени, чем сотрудники с низким окладом

```python
# Оставляю в датафрейме нужные столбцы с уровнем зарплаты и среднем количество часов на рабочем месте в месяц
df_average_montly_hours = df[['average_montly_hours','salary']]
df_average_montly_hours
```

```python
#Создаю датафрейм с высокими зарплатами
df_high_salary = df_average_montly_hours[df_average_montly_hours['salary']=='high']
df_high_salary
```

```python
df_high_salary.describe()
```

```python
#Гистограмма распределения количества часов в датафрейме с высокими зарплатами
df_high_salary.plot(kind='hist', bins=10, alpha = 0.8)
```

```python
ax_high = sns.distplot(df_high_salary.average_montly_hours)
```

```python
#Среднее количество часов в датафрейме с высокими зарплатами
df_high_salary['average_montly_hours'].mean()
```

```python
#Дисперсия количества часов в датафрейме с высокими зарплатами
df_high_salary['average_montly_hours'].var()
```

```python
#Создаю датафрейм с низкими зарплатами
df_low_salary = df_average_montly_hours[df_average_montly_hours['salary']=='low']
df_low_salary
```

```python
df_low_salary.describe()
```

```python
#Гистограмма распределения количества часов в датафрейме с низкими зарплатами
df_low_salary.plot(kind='hist', bins=6, alpha = 0.8)
```

```python
ax_low = sns.distplot(df_low_salary.average_montly_hours)
```

```python
#Среднее количество часов в датафрейме с низкими зарплатами
df_low_salary['average_montly_hours'].mean()
```

```python
#Дисперсия количества часов в датафрейме с низкими зарплатами
df_low_salary['average_montly_hours'].var()
```

```python
#Разница дисперсий
(df_low_salary['average_montly_hours'].var()-df_high_salary['average_montly_hours'].var())/df_high_salary['average_montly_hours'].var()*100
```

Сравнивая средние по датафреймам с высокими и низкими зарплатами видно, что величина в датафрейме с высокоми зарплатами (199.86742118027485) чуть ниже, чем в датафреме с низкими (200.9965828321487). Предварительно понятно, что сотрудники с высоким окладом проводят на работе не больше времени, чем сотрудники с низким окладом.

Формулирую нулевую и альтернативную гипотезы:
H0- сотрудники с высоким окладом проводят на работе столько же времени, сколько и сотрудники с низким окладом
H1 - сотрудники с высоким окладом и сотрудники с низким окладом проводят на работе разное количество времени

Пороговый уровень статистической значимости - 0,05

Нужен двухвыборочный двухсторонний критерий.

Так как дисперсии выборок отличаются при использовании критерия Стьюдента буду использовать критерия Уэлча (equal_var=False).

```python
#Проверка гипотезы с помощью критерия Стьюдента
from scipy import stats as st
alpha = 0.05
result = st.ttest_ind(df_high_salary['average_montly_hours'], df_low_salary['average_montly_hours'], equal_var=False)
print(result)

if (result.pvalue < alpha):
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу - сотрудники с высоким окладом проводят на работе столько же времени, сколько и сотрудники с низким окладом')
```

```python
#Среднее время,которое ежемесячно проводят на работе низкооплачиваемые сотрудники
df_low_salary['average_montly_hours'].mean()
```

```python
#Среднее время,которое ежемесячно проводят на работе высокооплачиваемые сотрудники
df_high_salary['average_montly_hours'].mean()
```

Различие между этими величинами (средними) статистически незначимо, не можем утверждать, что сотрудники с высоким окладом проводят на работе больше времени, чем сотрудники с низким окладом.

```python
#Проверю еще гипотезы с помощью критерия Манна-Уитни (так как данные все же не имеют нормального распределения и разница в дисперсиях более 5 %)
from scipy.stats import mannwhitneyu
alpha = 0.05
stat, p = mannwhitneyu(df_high_salary['average_montly_hours'], df_low_salary['average_montly_hours'])
print(f'Statictis = {stat}, p= {p}')
if p < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу - сотрудники с высоким окладом проводят на работе столько же времени, сколько и сотрудники с низким окладом')
```

Вторая проверка подтвердила предыдущий результат - мы не можем утверждать, что сотрудники с высоким окладом проводят на работе больше времени, чем сотрудники с низким окладом.

## **LDA**

```python
#Оставляю в датафрейме нужные факторы(колонки), задаю целевую переменную 'left'
data = df[['satisfaction_level', 'last_evaluation', 'number_project', 'average_montly_hours', 'time_spend_company','Work_accident', 'left']]
data_class = df['left'].to_numpy()
```

```python
#Значения целевой переменной
data_class
```

```python
# Разделяю выборку на тренировочную и тестовую
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data, data_class, test_size=0.3)
```

```python
len(y_test)
```

```python
y_test
```

```python
x_test
```

```python
# импортирую LDA-функционал
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis()
```

```python
# обучаю данные
lda.fit(x_train, y_train)
```

```python
# делаю прогноз на тестовой выборке
y = lda.predict(x_test)
print(type(y))
print(y[0:40])
```

```python
y_test
```

```python
# Разница факта и прогноза
result = pd.DataFrame([y_test, lda.predict(x_test)]).T
result[0:40]
```

```python
# расчет точности - отношение верный прогнозов к общему количеству позиций
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, lda.predict(x_test))
```

Модель не очень точная, на другой выборке получались бОльшие значения этого показателя.

```python
#Так как результат отличается в зависимости от случайного разбиения данных на train и test
#для уточнения точности модели применяю один из механизмов кросс-валидации
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import cross_val_score
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)

scores = cross_val_score(lda, data, data_class, scoring='accuracy', cv=cv, n_jobs=-1)
print(np.mean(scores)) 
```

Изменения незначительны в показателе.
