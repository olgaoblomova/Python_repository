```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
#–ë—É–¥—É –∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å —Å—á–∞—Å—Ç—å–µ!
df_2015 = pd.read_csv('2015.csv')
df_2016 = pd.read_csv('2016.csv')
df_2017 = pd.read_csv('2017.csv')
df_2018 = pd.read_csv('2018.csv')
df_2019 = pd.read_csv('2019.csv')
df_2015
```

```python
#–î–æ–±–∞–≤–ª—è—é –≤ –∫–∞–∂–¥—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å—Ç–æ–ª–±–µ—Ü —Å —É–∫–∞–∑–∞–Ω–∏–µ–º –≥–æ–¥–∞
df_2015.loc[:, 'year'] = '2015'
df_2015
df_2016.loc[:, 'year'] = '2016'
df_2016
df_2017.loc[:, 'year'] = '2017'
df_2017
df_2018.loc[:, 'year'] = '2018'
df_2018
df_2019.loc[:, 'year'] = '2019'
df_2019
```

```python
#–ü—Ä–∏–≤–æ–∂—É –¥–∞—Ç–∞—Ñ—Ä–µ–π–º—ã –∫ –µ–¥–∏–Ω–æ–æ–±—Ä–∞–∑–∏—é
df_2015_for_unit = df_2015[['year','Country', 'Region', 'Happiness Rank', 'Happiness Score', 'Economy (GDP per Capita)', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Standard Error', 'Family',  'Generosity', 'Dystopia Residual']]
df_2015_for_unit.rename(columns={'Happiness Rank': 'Overall rank', 'Economy (GDP per Capita)': 'GDP per capita', 'Health (Life Expectancy)':'Healthy life expectancy', 'Trust (Government Corruption)': 'Perceptions of corruption'}, inplace=True)
df_2015_for_unit
df_2016_for_unit = df_2016[['year','Country', 'Region', 'Happiness Rank', 'Happiness Score', 'Economy (GDP per Capita)', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Lower Confidence Interval', 'Upper Confidence Interval', 'Family', 'Generosity',
       'Dystopia Residual']]
df_2016_for_unit.rename(columns={'Happiness Rank': 'Overall rank', 'Economy (GDP per Capita)': 'GDP per capita', 'Health (Life Expectancy)':'Healthy life expectancy', 'Trust (Government Corruption)': 'Perceptions of corruption'}, inplace=True)
df_2016_for_unit
df_2017_for_unit = df_2017[['year','Country', 'Happiness.Rank', 'Happiness.Score', 'Economy..GDP.per.Capita.', 'Health..Life.Expectancy.', 'Freedom', 'Trust..Government.Corruption.', 'Generosity','Whisker.high', 'Whisker.low', 'Family','Dystopia.Residual']]
df_2017_for_unit.rename(columns={'Happiness.Rank': 'Overall rank', 'Happiness.Score': 'Happiness Score', 'Economy..GDP.per.Capita.': 'GDP per capita', 'Health..Life.Expectancy.':'Healthy life expectancy', 'Trust..Government.Corruption.': 'Perceptions of corruption'}, inplace=True)
df_2017_for_unit
df_2018_for_unit = df_2018[['year','Country or region', 'Overall rank', 'Score', 'GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']]
df_2018_for_unit.rename(columns={'Country or region': 'Country', 'Score': 'Happiness Score', 'Freedom to make life choices': 'Freedom'}, inplace=True)

df_2018_for_unit
df_2019_for_unit = df_2019[['year','Country or region', 'Overall rank', 'Score', 'GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']]
df_2019_for_unit
df_2019_for_unit.rename(columns={'Country or region': 'Country', 'Score': 'Happiness Score', 'Freedom to make life choices': 'Freedom'}, inplace=True)
df_2019_for_unit
```

```python
#–û–±—ä–µ–¥–∏–Ω—è—é –¥–∞—Ç–∞—Ñ—Ä–µ–π–º—ã –≤ –æ–¥–∏–Ω
df_unit = pd.concat([df_2015_for_unit,df_2016_for_unit, df_2017_for_unit, df_2018_for_unit, df_2019_for_unit])
df_unit.info()
```
