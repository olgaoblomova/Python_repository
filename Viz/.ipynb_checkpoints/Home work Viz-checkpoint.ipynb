```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
#Буду анализировать счастье!
df_2015 = pd.read_csv('2015.csv')
df_2016 = pd.read_csv('2016.csv')
df_2017 = pd.read_csv('2017.csv')
df_2018 = pd.read_csv('2018.csv')
df_2019 = pd.read_csv('2019.csv')
df_2015
```

```python
df_2015.describe()
```

```python
#Добавляю в каждый датафрейм столбец с указанием года
df_2015.loc[:, 'year'] = '2015'
df_2015
df_2016.loc[:, 'year'] = '2016'
df_2016
df_2017.loc[:, 'year'] = '2017'
df_2017
df_2018.loc[:, 'year'] = '2018'
df_2018
df_2019.loc[:, 'year'] = '2019'
df_2019
```

```python
df_2016.head()
```

```python
df_2019.columns
```

```python
#Привожу датафреймы к единообразию
df_2015_for_unit = df_2015[['year','Country', 'Region', 'Happiness Rank', 'Happiness Score', 'Economy (GDP per Capita)', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Standard Error', 'Family',  'Generosity', 'Dystopia Residual']]
df_2015_for_unit.rename(columns={'Happiness Rank': 'Overall rank', 'Economy (GDP per Capita)': 'GDP per capita', 'Health (Life Expectancy)':'Healthy life expectancy', 'Trust (Government Corruption)': 'Perceptions of corruption'}, inplace=True)
df_2015_for_unit
df_2016_for_unit = df_2016[['year','Country', 'Region', 'Happiness Rank', 'Happiness Score', 'Economy (GDP per Capita)', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Lower Confidence Interval', 'Upper Confidence Interval', 'Family', 'Generosity',
       'Dystopia Residual']]
df_2016_for_unit.rename(columns={'Happiness Rank': 'Overall rank', 'Economy (GDP per Capita)': 'GDP per capita', 'Health (Life Expectancy)':'Healthy life expectancy', 'Trust (Government Corruption)': 'Perceptions of corruption'}, inplace=True)
df_2016_for_unit
df_2017_for_unit = df_2017[['year','Country', 'Happiness.Rank', 'Happiness.Score', 'Economy..GDP.per.Capita.', 'Health..Life.Expectancy.', 'Freedom', 'Trust..Government.Corruption.', 'Generosity','Whisker.high', 'Whisker.low', 'Family','Dystopia.Residual']]
df_2017_for_unit.rename(columns={'Happiness.Rank': 'Overall rank', 'Happiness.Score': 'Happiness Score', 'Economy..GDP.per.Capita.': 'GDP per capita', 'Health..Life.Expectancy.':'Healthy life expectancy', 'Trust..Government.Corruption.': 'Perceptions of corruption'}, inplace=True)
df_2017_for_unit
df_2018_for_unit = df_2018[['year','Country or region', 'Overall rank', 'Score', 'GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']]
df_2018_for_unit.rename(columns={'Country or region': 'Country', 'Score': 'Happiness Score', 'Freedom to make life choices': 'Freedom'}, inplace=True)

df_2018_for_unit
df_2019_for_unit = df_2019[['year','Country or region', 'Overall rank', 'Score', 'GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']]
df_2019_for_unit
df_2019_for_unit.rename(columns={'Country or region': 'Country', 'Score': 'Happiness Score', 'Freedom to make life choices': 'Freedom'}, inplace=True)
df_2019_for_unit
```

```python
#Объединяю датафреймы в один
df_unit = pd.concat([df_2015_for_unit,df_2016_for_unit, df_2017_for_unit, df_2018_for_unit, df_2019_for_unit])
df_unit.info()
```

👀👀Посмотрю предварительно созависимости между столбцами, по которым есть данные за все годы

```python
sns.pairplot(df_unit[['Country','Happiness Score', 'Healthy life expectancy', 'GDP per capita', 'Freedom', 'Generosity','year']])
```

Наиболее ярко выражена созависимость между уровнем счастья и ВВП, и уровнем счастья и продолжительностью жизни. Есть некая зависимость от уровня свободы, не такая выраженная, как от двух предыдущих показателей. Вот от щедрости я завимости индекса счастья не увидела. Явно есть зависимость между продолжительностью жизни и уровнем ВВП.

👀👀Построю гистограмму распределения Индекса счастья по годам. Разобъю на 6 интервалов.


```python
print ('Распределение индекса счастья')
for title, group in df_unit[['year', 'Happiness Score']].groupby('year'):
    group.plot(kind = 'hist', title = title,  bins=6,
                   xlabel='Индекс счастья',
                   ylabel='Частота',
                 xticks=range(1, 8),
                alpha = 0.8) 
    plt.axis([2,8,0,50])
plt.show()                
```

👀👀Наиболее часто встречающийся индекс счастья лежит в интервале от около 5 до 6. Страны с индеком около 4 и чуть более 6 встречаются с примерно одинаковой частотой. Стран с самым высоким индексом счастья больше, чем стран с самым низким уровнем этого показателя. Распределение индекса счастья менялось от года к году. Так в 2015 году больше всего (более 40) значений было с индексом около 5. В 2016 году примерно одинаковое количество показателя счастья со значением от 4 до 6, но количественно менее 40. В 2017 году больше всего (также более 40) количество значений с увеличившимся индексом около 6. С 2015 по 2017 год очень мало значений индекса около 3. В 2018-2019 гг это значение увеличилось, но снизилось количество показателей со значением 5-6.

```python
#График созависимости между уровнем ВВП и уровнем счастья в разрезе годов
print('Графики оценки счастья и ВВП по странам')
for title, group in df_unit[['year', 'GDP per capita', 'Happiness Score']].groupby('year'):
    group.plot(kind = 'scatter', x = 'GDP per capita', y = 'Happiness Score', title = title )
    #plt.title('Графики оценки счастья и ВВП по странам')
```

👀👀Наблюдаю некую зависимость между показателем счастья и ВВП. По мере роста ВВП страны растет и ее оценка счастья. Максимальная оценка счастья достигается при высоком, но не самом высоком уровне ВВП, при максимальных ВВП оценка счастья находится между 6 и 7. В 2018 году по одной стране есть выброс (это Арабские Эмираты) -уровень счастья около 7 при максимальном значении ВВП по датафрейму (2.096).

```python
#График созависимости между продолжительностью жизни  и уровнем счастья в разрезе годов
print('Соотношение между продолжительностью жизни и индексом счастья')
for title, group in df_unit[['year', 'Healthy life expectancy', 'Happiness Score']].groupby('year'):
    group.plot(kind = 'scatter', x = 'Healthy life expectancy', y = 'Happiness Score', title = title )
```

В целом, при наиболее высокой продолжительности жизни уровень счастья выше. С 2015 по 2018 гг прослеживаются значения уровня счастья 4-5 при продолжительности жизни около 10 лет, думаю, здесь ошибки в данных. 

В 2019 году выросло по сравнению с 2015 годом количество стран, в которых показатель продолжительности жизни превышает 1 (100 лет?). И показатель стал более требовательным к продолжительности жизни. Например, в 2015 году оценка счастья достигала 7 при значениях показателя 0,70-0,90, а в 2019 году та же оценка достигается при значениях около 1 и более. 


Средняя продолжительность жизни увеличилась в 2019 по сравнению с 2015 годом.
