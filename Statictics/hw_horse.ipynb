```python
def analysis_column(colunm):
    """
    –§—É–Ω–∫—Ü–∏—è –≤—ã–≤–æ–¥–∏—Ç –Ω–∞ —ç–∫—Ä–∞–Ω –≤—Å–µ –±–∞–∑–æ–≤—ã–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏, –≥—Ä–∞–Ω–∏—Ü—ã –≤—ã–±—Ä–æ—Å–æ–≤, –ø—É—Å—Ç—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞
    """
    print(f'–ê–Ω–∞–ª–∏–∑–∏—Ä—É–µ–º —Å—Ç–æ–ª–±–µ—Ü {colunm}:')
    print('*'*30)
    #–°—Ä–µ–¥–Ω–∏–µ
    print('–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ = ', df_8[colunm].max())
    print('–ú–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ ', df_8[colunm].min())
    print('–°—Ä–µ–¥–Ω–µ–∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ = ', df_8[colunm].mean())
    print('–ú–æ–¥–∞ = ', df_8[colunm].mode()[0])
    print('–ú–µ–¥–∏–∞–Ω–∞ = ', df_8[colunm].median())
    ########################
    print('–°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ = ', df_8[colunm].std())
    print('–î–∏—Å–ø–µ—Ä—Å–∏—è =  ', df_8[colunm].var())
    print('0,25 –∏ 0,75 –∫–≤–∞–Ω—Ç–∏–ª–∏:', df_8[colunm].quantile([0.25, 0.75]))
    # –º–µ–∂–∫–≤–∞—Ä—Ç–∏–ª—å–Ω—ã–π —Ä–∞–∑–º–∞—Ö
    Q1 = df_8[colunm].quantile(0.25)
    print('Q1 = ', Q1)
    Q3 = df_8[colunm].quantile(0.75)
    print('Q3 = ', Q3)
    IQR = Q3 - Q1
    print('IQR = ', IQR)

    #–í—ã–±—Ä–æ—Å—ã
    print('–ê–Ω–∞–ª–∏–∑–∏—Ä—É–µ–º –≤—ã–±—Ä–æ—Å—ã:')
    print('*'*30)
    print('–£–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–∞:', sorted(df_8[colunm].unique()))
    lower_bound = Q1 - (1.5 * IQR) 
    upper_bound = Q3 + (1.5 * IQR)
    print('lower_outliers = ', lower_bound)
    print('upper_outliers = ', upper_bound)
    print('–ê–Ω–∞–ª–∏–∑–∏—Ä—É–µ–º –≤–µ—Ä—Ö–Ω—é—é –≥—Ä–∞–Ω–∏—Ü—É:')
    print('*'*30)
    print(df_8[df_8[colunm] > upper_bound])
    print('–ê–Ω–∞–ª–∏–∑–∏—Ä—É–µ–º –Ω–∏–∂–Ω—é—é –≥—Ä–∞–Ω–∏—Ü—É:')
    print('*'*30)
    print(df_8[df_8[colunm] < lower_bound])

    print('–ü—Ä–æ–ø—É—Å–∫–∏:')
    print('*'*30)
    print(df_8[df_8[colunm].isnull()])  
    print('–ü—É—Å—Ç—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π:', len(df_8[df_8[colunm].isnull()]))
```
