```python
def analysis_column(colunm):
    """
    Функция выводит на экран все базовые статистики, границы выбросов, пустые значения заданного столбца
    """
    print(f'Анализируем столбец {colunm}:')
    print('*'*30)
    #Средние
    print('Максимальное значение = ', df_8[colunm].max())
    print('Минимальное значение ', df_8[colunm].min())
    print('Среднеарифметическое значение = ', df_8[colunm].mean())
    print('Мода = ', df_8[colunm].mode()[0])
    print('Медиана = ', df_8[colunm].median())
    ########################
    print('Стандартное отклонение = ', df_8[colunm].std())
    print('Дисперсия =  ', df_8[colunm].var())
    print('0,25 и 0,75 квантили:', df_8[colunm].quantile([0.25, 0.75]))
    # межквартильный размах
    Q1 = df_8[colunm].quantile(0.25)
    print('Q1 = ', Q1)
    Q3 = df_8[colunm].quantile(0.75)
    print('Q3 = ', Q3)
    IQR = Q3 - Q1
    print('IQR = ', IQR)

    #Выбросы
    print('Анализируем выбросы:')
    print('*'*30)
    print('Уникальные значения столбца:', sorted(df_8[colunm].unique()))
    lower_bound = Q1 - (1.5 * IQR) 
    upper_bound = Q3 + (1.5 * IQR)
    print('lower_outliers = ', lower_bound)
    print('upper_outliers = ', upper_bound)
    print('Анализируем верхнюю границу:')
    print('*'*30)
    print(df_8[df_8[colunm] > upper_bound])
    print('Анализируем нижнюю границу:')
    print('*'*30)
    print(df_8[df_8[colunm] < lower_bound])

    print('Пропуски:')
    print('*'*30)
    print(df_8[df_8[colunm].isnull()])  
    print('Пустых значений:', len(df_8[df_8[colunm].isnull()]))
```
