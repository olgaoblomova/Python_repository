```python
# Изучаю возраст лошадей
#Средние
print('Максимальный возраст лошади = ', df_8['Age'].max(), 'лет/год(а)')
print('Минимальный возраст лошади = ', df_8['Age'].min(), 'лет/год(а)')
print('Среднеарифметический возраст лошади = ', df_8['Age'].mean(), 'лет/год(а)')
print('Самый часто встречающийся возраст лошади = ', df_8['Age'].mode()[0], 'лет/год(а)')
print('Медианный возраст лошади = ', df_8['Age'].median(), 'лет/год(а)')
########################
print('Стандартное отклонение = ', df_8['Age'].std())
print('Дисперсия =  ', df_8['Age'].var())
print('0,25 и 0,75 квантили:', df_8['Age'].quantile([0.25, 0.75]))
# межквартильный размах
Q1 = df_8['Age'].quantile(0.25)
print('Q1 = ', Q1)
Q3 = df_8['Age'].quantile(0.75)
print('Q3 = ', Q1)
IQR = Q3 - Q1
print('IQR = ', IQR)

#Выбросы
print('Уникальные значения столбца:', df_8['Age'].unique())
len(df_8.loc[df_8.Age == 9])
# Если исключить выборсы среднеарифметический возраст несколько изменится с 1,64 до 1 года, медиана и мода нет.
lower_bound = Q1 - (1.5 * IQR) 
upper_bound = Q3 + (1.5 * IQR)
print('lower_bound = ', lower_bound)
print('upper_bound = ', upper_bound)
# remove_outliers = df_8[df_8['Age'].between(lower_bound, upper_bound, inclusive=True)]
# remove_outliers
```
