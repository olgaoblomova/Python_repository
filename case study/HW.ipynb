## Критикам нравятся спортивные игры?
Будем считать, что критикам нравятся спортивные игры, если их оценки превышают 7 баллов.
Критики оценивают спортивные игры **больше**, чем на 7 баллов? (односторонний тест)

H0: средняя оценка критиков спортивных игр  <= 7 баллов
H1: средняя оценка критиков спортивных игр  более 7 баллов

За alpha здесь и далее будем принимать 0.05

Сравниваем значение по выборке с константой, поэтому нужен одновыборочный критерий.

```python
#Использование параметра nan_policy  = 'omit' к столбцу содержащему None выдает ошибку
from scipy import stats as st
alpha = 0.05
result = st.ttest_1samp(df_sport['Critic_Score'], 7.0, nan_policy  = 'omit', alternative ='greater')

print(df_sport['Critic_Score'].mean())
print(result)

if result.pvalue < alpha: 
    print('Отвергаем нулевую гипотезу, средняя оценка критиков спортивных игр  более 7 баллов')
else:
    print('Не отвергаем нулевую нулевую гипотезу, средняя оценка критиков спортивных игр  <= 7 баллов')
```

```python
from scipy import stats as st
alpha = 0.05
result = st.ttest_1samp(dfff, 7.0, alternative ='greater')

print(dfff.mean())
print(result)

if result.pvalue < alpha: 
    print('Отвергаем нулевую гипотезу, средняя оценка критиков спортивных игр  более 7 баллов')
else:
    print('Не отвергаем нулевую нулевую гипотезу, средняя оценка критиков спортивных игр  <= 7 баллов')
```

## Критикам больше нравятся игры на PC чем на PS4? (двусторонний тест)

Н0:Критикам одинаково нравятся игры на PC и на PS4
H1:Критикам по-разному нравятся игры на PC и на PS4 (исходя из средних величин на PS4 нравятся больше, на PC)

Нужен двухвыборочный двухсторонний критерий.

Так как дисперсии выборок отличаются при использовании критерия Стьюдента буду использовать критерия Уэлча (equal_var=False)

```python
#Проверка гипотезы с помощью критерия Стьюдента
from scipy import stats as st
alpha = 0.05
result = st.ttest_ind(df_pc['Critic_Score'], df_ps4['Critic_Score'], nan_policy  = 'omit', equal_var=False)
print(result)

if (result.pvalue < alpha):
    print('Отвергаем нулевую гипотезу - Критикам по-разному нравятся игры на PC и на PS4 (исходя из средних величин на PS4 нравятся больше, на PC)')
else:
    print('Не отвергаем нулевую гипотезу - Критикам одинаково нравятся игры на PC и на PS4')
```

```python
#Проверю гипотезы с помощью критерия Манна-Уитни (так как данные все же не имеют нормального распределения и разница в дисперсиях более 5 %)
from scipy.stats import mannwhitneyu
alpha = 0.05
stat, p = mannwhitneyu(df_pc['Critic_Score'], df_ps4['Critic_Score'])
print(f'Statictis = {stat}, p= {p}')
if p < alpha:
    print('Отвергаем нулевую гипотезу - Критикам по-разному нравятся игры на PC и на PS4 (исходя из средних величин на PS4 нравятся больше, на PC)')
else:
    print('Не отвергаем нулевую гипотезу - Критикам одинаково нравятся игры на PC и на PS4')
```
